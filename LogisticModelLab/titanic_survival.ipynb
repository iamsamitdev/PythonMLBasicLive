#### Step 2: Connect to MySQL Database

```python
import mysql.connector

# Connect to the database
db_connection = mysql.connector.connect(
    host="localhost",
    user="root",
    password="123456",
    database="pythonmldb",
    port="3306"
)

# test the connection
print(db_connection)
```

#### Step 3: Fetching Data from MySQL Database to Pandas DataFrame

```python
import pandas as pd

# SQL query
query = "SELECT * FROM titanic"

# Load data into a DataFrame df from the MySQL query
titanic_df = pd.read_sql(query, con=db_connection)

# Display the DataFrame
titanic_df
```

#### Step 4: Data Preprocessing and Cleaning

```python
# Drop NaN values
titanic_df = titanic_df[['Age','SibSp','Parch','Fare','Survived']].dropna()

# Display the DataFrame
# titanic_df

# Split the data into features and target
x = titanic_df[['Age','SibSp','Parch','Fare']]
y = titanic_df['Survived']

print(x)
print(y)
```

#### Step 5: Spit Data into Training and Testing Sets

```python
from sklearn.model_selection import train_test_split

# Split the data into a training and a test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=0)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

#### Step 6: Scalling the Data

```python
from sklearn.preprocessing import StandardScaler

# Scale the data
scaler = StandardScaler()

# Fit the scaler to the training data and transform the training data
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

print(x_train)
print(x_test)
```

#### Step 7: Training the Model

```python
from sklearn.linear_model import LogisticRegression

# Create a logistic regression model
model = LogisticRegression()

# Train the model
model.fit(x_train, y_train)
```

#### Step 8: Predicting the Model

```python
y_pred = model.predict(x_test)

y_pred
```

#### Step 9: Model Evaluation (การประเมินผล)

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Accuracy
print('Accuracy:', accuracy_score(y_test, y_pred))

# Precision
print('Precision:', precision_score(y_test, y_pred))

# Recall
print('Recall:', recall_score(y_test, y_pred))

# F1 score
print('F1 score:', f1_score(y_test, y_pred))
```

#### Step 10: Export Model

```python
from joblib import dump

# Save the model
dump(model, 'titanic_model.joblib')
```

#### Step 11: Test use Model from joblib

```python
from joblib import load

# Load the model
loaded_model = load('titanic_model.joblib')

# Test the model
y_pred = loaded_model.predict([[22, 1, 0, 7.25]])

print(y_pred)
```
