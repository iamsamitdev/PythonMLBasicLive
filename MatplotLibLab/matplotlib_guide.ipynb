#### 2. การวาดกราฟเส้น (Line Graph)

```python
import matplotlib.pyplot as plt
import numpy as np

# กำหนดแกน x และ y ด้วย numpy array
x = np.array([2018, 2019, 2020, 2021, 2022])
y = np.array([1050, 2452, 3456, 4653, 4052])

# สร้างกราฟด้วย matplotlib
plt.plot(x, y)
plt.xticks(x)
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

#### 3. การวาดกราฟแท่ง (Bar Graph)

```python
plt.bar(x, y)
plt.xticks(x)
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

#### 4. การโหลดจาก csv มา plot กราฟ

```python
import matplotlib.pyplot as plt
import pandas as pd

# สร้าง DataFrame จากไฟล์ CSV
df = pd.read_csv('datasets/gdp.csv')

# สร้างกราฟเส้นด้วย matplotlib
plt.plot(df['year'], df['gdp']) # สร้างเส้นกราฟ
plt.xticks(df['year']) # กำหนดแกน x ด้วยค่าในคอลัมน์ year
plt.title('Thailand GDP') # กำหนดชื่อกราฟ
plt.xlabel('Year') # กำหนดชื่อแกน x
plt.ylabel('GDP (Billion Baht)') # กำหนดชื่อแกน y
plt.show() # แสดงกราฟ
```

```python
# bar chart
plt.bar(
    df['year'], df['gdp'], 
    color=['green', 'red', 'blue', 'yellow', 'pink'], 
    edgecolor='black'
)
plt.xticks(df['year']) # กำหนดแกน x ด้วยค่าในคอลัมน์ year
plt.title('Thailand GDP') # กำหนดชื่อกราฟ
plt.xlabel('Year') # กำหนดชื่อแกน x
plt.ylabel('GDP (Billion Baht)') # กำหนดชื่อแกน y
plt.show() # แสดงกราฟ
```

```python
# pie chart
plt.pie(
    df['gdp'], 
    labels=df['year'], 
    autopct='%.2f%%', 
    colors=['green', 'red', 'blue', 'yellow', 'pink']
)
plt.title('Thailand GDP') # กำหนดชื่อกราฟ
plt.show() # แสดงกราฟ
```
