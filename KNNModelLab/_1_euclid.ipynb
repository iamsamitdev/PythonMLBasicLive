#### 1.Plot the data

```python
import matplotlib.pyplot as plt

# สร้าง list สำหรับการแสดงผลบน x และ y
x = [3, 8, 14, 3]
y = [4, 13, 5, 4]

# Plot the points
plt.plot(x, y, linestyle='-', marker='o', color='r')

# แสดง label และ title
plt.xlabel('X')
plt.ylabel('Y')

# แสดงจุดที่ใช้ plot
point = ['A', 'B', 'C', 'A']

# แสดง label ของจุด
for i in range(len(x)):
    plt.text(x[i], y[i], f'{point[i]} ({x[i]},{y[i]})')

# แสดง title
plt.title('Distance between point A, B, and C')

# แสดงผลกราฟ
plt.show()
```

#### 2. คำนวณหาระยะทางระหว่างจุดทั้งหมด โดยใช้สูตร Euclidean distance

```python
from sklearn.metrics.pairwise import euclidean_distances

# กำหนดจุด A, B, C
a = [[3, 4]]
b = [[8, 13]]
c = [[14, 5]]

# คำนวณระยะห่างระหว่างจุด A และ B และ C
ab = euclidean_distances(a, b)
bc = euclidean_distances(b, c)
ca = euclidean_distances(c, a)

# แสดงผลลัพธ์
print(f'ระยะห่างระหว่างจุด A และ B: {ab[0][0]:.3f}')
print(f'ระยะห่างระหว่างจุด B และ C: {bc[0][0]:.3f}')
print(f'ระยะห่างระหว่างจุด C และ A: {ca[0][0]:.3f}')
```
